Створити візуалізацію, аналогічно рисунку 
- для колонки `a` використайте червоний колір та лінію формату `-.`

- для колонки `b` використайте помаранчевий колір та товщину `10`

- для колонки `c` використайте жовтий колір та товщину `1` і маркер `o`


```python
sns.set_style('darkgrid')
fig, axes = plt.subplots(figsize=(15,8))
x=df.index
y1=df['a']
y2=df['b']
y3=df['c']
axes.plot(x, y1, c="r", ls='-.')
axes.plot(x, y2, c="orange", lw=10)
axes.plot(x, y3, c="yellow", lw=1, marker='o')
```

Створити візуалізацію, аналогічно рисунку 
- і не забудьте про легенду :)

```python
sns.set_style('darkgrid')
fig, axes = plt.subplots(3, figsize=(15,8))
x=df.index
y1=df['a']
y2=df['b']
y3=df['c']
axes[0].plot(x, y1, c="tab:blue", label="Line A")
axes[1].plot(x, y2, c="tab:blue", label="Line B")
axes[2].plot(x, y3, c="tab:blue")
axes[0].legend(loc=1)
axes[1].legend(loc=3)   
```

Створити візуалізацію, аналогічно рисунку 
- використайте томатний колір та відстань між стовбцями 0.5

```python
sns.set_style('darkgrid')
fig, axes = plt.subplots(figsize=(15,8))
x=df.index
y1=df['a']
axes.bar(x, y1, align="center", width=0.8, alpha=0.6, label="a", color="tomato")
axes.legend(loc=1)
```

Створити візуалізацію, аналогічно рисунку 
- добавте всі підписи та правильний маркер

```python
sns.set_style('darkgrid')
fig, axes = plt.subplots(figsize=(17,8))

axes.set_title("This is the Title")
axes.set_xlabel("This is the X Axis")
axes.set_ylabel("This is the Y Axis")
x=df.index
y=df['a']
axes.plot(x, y, c="tab:blue", marker='^')
```
